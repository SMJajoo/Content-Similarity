```python
import pdfplumber
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

def document_similarity(doc1, doc2):
    # Combine the documents into a list
    documents = [doc1, doc2]
    
    # Create a TF-IDF Vectorizer
    vectorizer = TfidfVectorizer()
    
    # Fit and transform the documents into TF-IDF vectors
    tfidf_matrix = vectorizer.fit_transform(documents)
    # Calculate the cosine similarity between the vectors
    cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    
    return cosine_sim[0][0]

# Paths to the two PDF files
pdf_file1 = '/kaggle/input/document-similarity/SQL Cheatsheet for interviews.pdf'
pdf_file2 = '/kaggle/input/document-similarity/SQL joins.pdf'

# Extract text from both PDFs
doc1_text = extract_text_from_pdf(pdf_file1)
doc2_text = extract_text_from_pdf(pdf_file2)

# Calculate the similarity
similarity_score = document_similarity(doc1_text, doc2_text)

print(f"Similarity Score: {similarity_score}")
```
